```python
def func(s):
    try:
        float(s)
        return True
    except ValueError:
        return False
        
print(func('26587'))
print(func('4.2365'))
print(func('-12547'))
print(func('00'))
print(func('A001'))
print(func('001'))
```

```python
def is_even(n):
    return n%2 == 0 
print(is_even(6))  
print(is_even(7))

nums = [1, 2, 3, 4, 5]
squares = list(map(lambda x: x**2, nums))
print(squares)

odds = list(filter(lambda x: x % 2 == 1, nums))
print(odds) # [1, 3, 5]
```
