```python
import numpy as np

def f_to_c(F):
    return (F-32)*5/9
def c_to_f(C):
    return C*9/5+32

F = np.array([0, 12, 45.21, 34, 99.91, 32], dtype=float)

C = f_to_c(F)
C_rounded = np.round(C, 2) 

print("Фаренгейтийн градусын утгууд:", F)
print("Цельсийн градусын утгууд:", C_rounded)
```

```python
from pandas import DataFrame
import numpy as np

a = DataFrame(
    {
        'name':     ['Dorj', 'Khulan', 'Sergelen', 'Ganchimeg', 'Khaliun', 'Bolor', 'Anu', 'Tsetse', 'Purev', 'Zulaa'],
        'score':    [12.5, 9, 16.5, np.nan, 9, 20, 14.5, np.nan, 8, 19],
        'attempts': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
        'qualify':  ['yes', 'no', 'yes', 'no', 'no', 'yes', 'yes', 'no', 'no', 'yes']
    },
    index=['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']
)

print('Dasgal_1')
print(a.head(3))

print('Dasgal_2')
print(a[['name', 'score']])

print('Dasgal_3')
print(a[a['attempts'] > 2])

print('Dasgal_4')
print(a[a['score'].between(15, 20)])

print('Dasgal_5')
print(a.loc[['a', 'c', 'e', 'f'], ['name', 'score']])
```
