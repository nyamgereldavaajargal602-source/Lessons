```python
import numpy as np
from scipy.stats import norm, chi2

B = [
    24, 23, 24, 25, 29, 26, 21, 26, 20, 28, 28, 23, 29, 23, 11, 27, 14, 9,
    17, 25, 1, 39, 16, 28, 34, 27, 20, 16, 40, 18, 16, 19, 37, 8, 11, 28,
    32, 22, 16, 22, 18, 18, 19, 27, 16, 24, 18, 32, 26, 19, 9, 22, 24, 20,
    14, 32, 33, 15, 22, 4, 31, 18, 2, 26, 16, 14, 25, 22, 12, 19, 24, 10,
    25, 39, 23, 18, 14, 33, 34, 17, 20, 43, 29, 21, 20, 32, 22, 20, 28, 26,
    4, 9, 24, 24, 2, 32, 31, 33, 12, 17, 27, 27, 15, 23, 25, 10, 23, 11,
    21, 13, 23, 19, 20, 22, 29, 31, 31, 17, 19, 28, 12, 31, 5, 7, 19, 30,
    35, 15, 22, 18, 27, 8, 22, 15, 18, 22, 28, 31, 16, 24, 24, 28, 26, 15,
    3, 25, 4, 15, 25, 21, 16, 20, 4, 25, 6, 32, 32, 17, 20, 23, 31, 15,
    13, 41, 15, 39, 14, 15, 15, 12, 35, 12, 14, 29, 16, 26, 15, 22, 24, 14,
    18, 5, 27, 32, 21, 2, 30, 29, 36, 8, 19, 5, 24
]

x = np.array(B)
n = len(x)
alpha = 0.1

# 2. Дундаж ба Стандарт хазайлтыг тооцоолох
mean_val = np.mean(x)
std_val = np.std(x, ddof=1) # Засварласан стандарт хазайлт

# 3. Интервалын зайг 4-өөр авч, bins үүсгэх
# Хамгийн бага утгаас (1) эхлэн 4-ийн алхамтай интервалууд
h = 4
bins = np.arange(min(x), max(x) + h, h)

observed = []
expected = []

# 4. Интервал бүр дээрх давтамжийг тооцоолох
for i in range(len(bins) - 1):
    left = bins[i]
    right = bins[i + 1]
    
    # ni: Ажиглагдсан давтамж
    obs = np.sum((x >= left) & (x < right))
    
    # pi: Онолын магадлал
    p = norm.cdf(right, mean_val, std_val) - norm.cdf(left, mean_val, std_val)
    
    observed.append(obs)
    expected.append(n * p)

observed = np.array(observed)
expected = np.array(expected)

# 5. Хи-квадрат ажиглалтын утга
# (Хуваагч тэг болохоос сэргийлж маш бага тоо 1e-10 нэмэв)
chi_obs = np.sum((observed - expected) ** 2 / expected)

# 6. Чөлөөний зэрэг: df = k - 1 - 2 (mean, std үнэлсэн тул)
k = len(observed)
df = k - 1 - 2

# 7. Хи-квадрат онолын утга
chi_critical = chi2.ppf(1 - alpha, df)

# Үр дүнг хэвлэх
print(f"n = {n}")
print(f"mean = {mean_val:.4f}")
print(f"std = {std_val:.4f}")
print(f"df = {df}")

print("\nInterval | Observed (ni) | Expected (n*pi)")
print("-" * 40)
for i in range(k):
    print(f"[{bins[i]:2}, {bins[i+1]:2}) | {observed[i]:13} | {expected[i]:14.4f}")

print("-" * 40)
print(f"A17~1 (Chi_obs, ajillaltiin utga) = {chi_obs:.4f}")
print(f"A17~2 (Chi_crit, kritic utga) = {chi_critical:.4f}")

if chi_obs <= chi_critical:
    print("\nШийдвэр: H0-г зөвшөөрнө (Хэвийн тархалттай)")
else:
    print("\nШийдвэр: H0-г няцаана (Хэвийн тархалтгүй)")
```

```python
import math
import numpy as np
from scipy.stats import t

C_new = [
    70, 72, 73, 71, 76, 74,
    72, 65, 72, 70, 73, 75,
    68, 68, 71, 69, 71, 69,
    68, 68, 69, 75, 66, 71,
    69, 66, 65, 72, 70, 73,
    73, 73, 69, 74, 74, 79,
    78, 75, 70, 72, 77, 72,
    68, 63, 72, 70, 70, 70,
    65, 70, 76, 72, 72, 70,
    75, 70, 70, 69, 73, 78,
    72, 73, 76, 64, 78, 77,
    74, 68, 74, 71, 70, 67,
    74, 69, 75, 74, 75
]

k = 13
alpha = 0.005

# Эхний k ширхэг утга авах
sample = np.array(C_new[:k], dtype=float)

n = len(sample)
df = n - 1

# 1) Түүврийн дундаж
x_bar = sample.mean()

# 2) M = [x_bar - 4]
M = math.floor(x_bar - 4)

# 3) Засварласан стандарт хазайлт Sb
Sb = sample.std(ddof=1)

# 4) Түүврийн ажиглалтын утга t_аж
t_observed = (x_bar - M) / (Sb / math.sqrt(n))

# 5) Нэг талт критик утга (H1: a > M тул баруун талт)
# t(1-alpha, df)
t_critical = t.ppf(1 - alpha, df)

# =========================================================
# Хариу хэвлэх
# =========================================================

print(f"Сонгосон түүвэр (k={k}):", sample.tolist())
print(f"n = {n}, df = {df}")
print(f"x_bar (dundaj utga) = {x_bar:.4f}")
print(f"M (taamaglasan matmematic dundaj) = floor(x_bar - 4) = {M}")
print(f"Sb = {Sb:.4f}")

print("-" * 30)
print(f"A18~1 (Түүврийн ажиглалтын утга t_аж) = {t_observed:.4f}")
print(f"A18~2 (Таамаглалын критик утга t_кр) = {t_critical:.4f}")
print("-" * 30)

# Шийдвэр гаргах
if t_observed > t_critical:
    print("Шийдвэр: H0-г няцаана (a > M байна)")
else:
    print("Шийдвэр: H0-г няцаах үндэсгүй")
```
